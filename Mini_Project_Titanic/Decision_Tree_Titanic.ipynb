# Decision Tree on Titanic Datasset

```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df = df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
```

```python
df
```

```python
df.isnull().sum() #finding null values present in each column
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
sum(df.duplicated())
```

```python
df = df.drop_duplicates()
```

```python
sum(df.duplicated())
```

```python
df
```

```python
df = pd.get_dummies(df,columns=['Embarked'])
```

```python
df
```

```python
df['Sex'].replace(['male','female'],[0,1],inplace=True)
```

```python
df
```

```python
x = df.drop(['Survived'],axis=1)
y = df['Survived']
```

```python
x
```

```python
y
```

# Ploting Decision Tree 

```python
from sklearn import tree as t
```

```python
model = t.DecisionTreeClassifier(max_depth = 3)
```

```python
model.fit(x,y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (15,10) )
t.plot_tree(model)
```

```python
plt.figure(figsize = (15,10) )
t.plot_tree(model,filled=True)
```

```python
model.score(x,y)
```
